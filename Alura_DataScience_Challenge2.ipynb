#üìå Extrac√£o

O formato do arquivo que precisamos ler √© json. Usaremos a fun√ß√£o read_json() da livraria pandas para ler o arquivo.

```python
import pandas as pd

#url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
url = 'https://raw.githubusercontent.com/Angel-Arroyo-H/alura-data-science-challenge-2/refs/heads/main/TelecomX_Data.json'
dados = pd.read_json(url)
dados
```

Ap√≥s ler o arquivo observamos que temos informa√ß√µes aninhadas nos campos customer, phone, internet e account, normalizaremos os dados usando a fun√ß√£o pandas.json_normalize()

```python
pd.json_normalize(dados['customer'])
```

```python
pd.json_normalize(dados['phone'])
```

```python
pd.json_normalize(dados['internet'])
```

```python
pd.json_normalize(dados['account'])
```

Agora que temos uma melhor ideia da estrutura de nossos dados usaremos as livrarias request e json para ler os dados em formato json e transformaremos esses dados em um diccionario.

```python
import requests
import json

dados_json = requests.get(url)
dados_dict = json.loads(dados_json.text)
pd.DataFrame(dados_dict)
```

Agora normalizaremos o diccionario com os dados usando a fun√ß√£o json_normalize()

```python
dados_normalizados  = pd.json_normalize(dados_dict,sep='_')
dados_normalizados
```

### Churn
Come√ßaremos pela vari√°vel resposta **Churn**

```python
dados_normalizados['Churn'].value_counts(dropna=False)
```

```python
print(f"Temos {len([i for i in dados_normalizados['Churn'] if i not in ['Yes','No']])} valores diferentes de Yes e No, dado que esse campo ser√° usado como vari√°vel resposta precisaremos remover esses registros.")
```

```python
dados_tratados = dados_normalizados[dados_normalizados['Churn'].isin(['Yes','No'])].copy()
dados_tratados['Churn'] = dados_tratados['Churn'].map({'Yes':True,'No':False})
```

Antes de revisar as dem√°s vari√°veis podemos agrupar elas pelo tipo de vari√°vel que esperamos de acordo com a descri√ß√£o e revisando rapidamente os valores que elas adotam. Para isso podemos usar a fun√ß√£o unique().

```python
for i in dados_tratados.columns[2:]:
  print(f'\n{i}')
  print(f'Valores √∫nicos: {dados_tratados[i].unique()}')
```

```python
variaveis_booleanas = [
    'customer_gender'
    ,'customer_SeniorCitizen'
    ,'customer_Partner'
    ,'customer_Dependents'
    ,'phone_PhoneService'
    ,'account_PaperlessBilling'
]
variaveis_categoricas = [
    'phone_MultipleLines'
    ,'internet_InternetService'
    ,'internet_OnlineSecurity'
    ,'internet_OnlineBackup'
    ,'internet_DeviceProtection'
    ,'internet_TechSupport'
    ,'internet_StreamingTV'
    ,'internet_StreamingMovies'
    ,'account_Contract'
    ,'account_PaymentMethod'
]
variaveis_int = [
    'customer_tenure'
]
variaveis_float = [
    'account_Charges_Monthly'
    ,'account_Charges_Total'
]
```

```python
# verificando que consideramos todas as vari√°veis
if [v for v in dados_tratados.columns[2:] if v not in variaveis_booleanas+variaveis_categoricas+variaveis_int+variaveis_float] == []:
  print('Todas as vari√°veis foram consideradas')
# verificando que n√£o temos interse√ß√µes entre grupos de vari√°veis
if len(set(variaveis_booleanas+variaveis_categoricas+variaveis_int+variaveis_float)) == len(variaveis_booleanas+variaveis_categoricas+variaveis_int+variaveis_float):
  print('Sem interse√ß√µes')
```

Agora revisamos se existem valores inconsistentes nas vari√°vies por tipo de vari√°vel.

### Vari√°veis booleanas:

```python
for i in variaveis_booleanas:
  print(f'\n')
  print(dados_tratados[i].value_counts(dropna=False))
```

N√£o temos valores inconsistentes nas vari√°veis booleanas, todas tem apenas 2 categorias. No entanto para treinar modelos eles apenas aceitar√°m valores num√©ricos ou booleanos, por tanto transformaremos os valores para o tipo bool. No caso da vari√°vel customer_gender criaremos uma vari√°vel equivalente que nomearemos customer_is_male.

```python
dados_tratados['customer_SeniorCitizen'] = dados_tratados['customer_SeniorCitizen'].map({0:False,1:True})
dados_tratados['customer_Partner'] = dados_tratados['customer_Partner'].map({'No':False,'Yes':True})
dados_tratados['customer_Dependents'] = dados_tratados['customer_Dependents'].map({'No':False,'Yes':True})
dados_tratados['phone_PhoneService'] = dados_tratados['phone_PhoneService'].map({'No':False,'Yes':True})
dados_tratados['account_PaperlessBilling'] = dados_tratados['account_PaperlessBilling'].map({'No':False,'Yes':True})
dados_tratados['customer_is_male'] = dados_tratados['customer_gender'].map({'Female':False,'Male':True})
dados_tratados.drop(columns='customer_gender',inplace=True) # removendo variavel costumer_gender

# Atualizamos a lista de vari√°veis booleanas
variaveis_booleanas.remove('customer_gender')
variaveis_booleanas.append('customer_is_male')
```

```python
# Verificando tipos de dados bool
dados_tratados[variaveis_booleanas].info()
```

### Variaveis int e float



```python
dados_tratados[variaveis_int+variaveis_float].info()
```

Precisaremos trocar o tipo de dados apenas da vari√°vel account_Charges_Total

```python
try:
  dados_tratados['account_Charges_Total'] = dados_tratados['account_Charges_Total'].astype(float)
except Exception as e:
  print(f"Ocorreu um erro tipo {type(e)}: {e}")
```

Isso significa que temos valores em branco no campo account_Charges_Total. Mas pode ser um espa√ßo em branco adicional ao n√∫mero ou que o dado tenha apenas o espa√ßo em branco. Precisamos identificar quais os elementos √∫nicos que est√°o gerando o erro.

```python
valores_com_espaco = [v for v in dados_tratados['account_Charges_Total'] if ' ' in v]
print(f'Quantidade de valores com espa√ßo em branco: {len(valores_com_espaco)} ({len(valores_com_espaco)/dados_tratados.shape[0]*100:.2f}%)')
print(f'Valores unicos com espa√ßo em branco: {set(valores_com_espaco)}')
```

```python
dados_tratados[dados_tratados['account_Charges_Total'] == ' ']
```

Por outra parte podemos ver que todos os registros onde account_Charges_Total = ' ' correspondem apenas √† categoria Churn = False. Considerando isso e o baixo numero de registros excluiremos esses casos e transformaremos a variavel para float.

```python
dados_tratados = dados_tratados[dados_tratados['account_Charges_Total'] != ' ']
dados_tratados['account_Charges_Total'] = dados_tratados['account_Charges_Total'].astype(float)
```

### Vari√°veis Categ√≥ricas

Revisaremos a quantidade de registros em cada categoria das vari√°veis.

```python
for i in variaveis_categoricas:
  print(f'\n')
  print(dados_tratados[i].value_counts(dropna=False))
```

N√£o observamos categorias inconsistentes com a descri√ß√£o das vari√°veis. No entanto podemos traduzir as categorias para o Portugu√™s, assim tornar os dados mais acess√≠veis.

```python
variaveis_categoricas
```

```python
dados_tratados['phone_MultipleLines'] = dados_tratados['phone_MultipleLines'].map({'No phone service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_InternetService'] = dados_tratados['internet_InternetService'].map({'Fiber optic':'Fibra √ìtica','No':'N√£o','DSL':'DSL'})
dados_tratados['internet_OnlineSecurity'] = dados_tratados['internet_OnlineSecurity'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_OnlineBackup'] = dados_tratados['internet_OnlineBackup'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_DeviceProtection'] = dados_tratados['internet_DeviceProtection'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_TechSupport'] = dados_tratados['internet_TechSupport'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_StreamingTV'] = dados_tratados['internet_StreamingTV'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['internet_StreamingMovies'] = dados_tratados['internet_StreamingMovies'].map({'No internet service':'Sem servi√ßo','No':'N√£o','Yes':'Sim'})
dados_tratados['account_Contract'] = dados_tratados['account_Contract'].map({'Month-to-month':'Mensal','One year':'Anual','Two year':'Dois anos'})
dados_tratados['account_PaymentMethod'] = dados_tratados['account_PaymentMethod'].map({
    'Electronic check':'Cheque eletr√¥nico'
    ,'Mailed check':'Cheque'
    ,'Bank transfer (automatic)':'Transfer√™ncia banc√°ria'
    ,'Credit card (automatic)':'Cart√£o de cr√©dito'
})
```

```python
for i in variaveis_categoricas:
  print(f'\n')
  print(dados_tratados[i].value_counts(dropna=False))
```

Agora que revisamos as inconsistencias e tratamos os dados podemos resetar o index e renomear as colunas para tornar a informa√ß√£o mais clara e acess√≠vel.
